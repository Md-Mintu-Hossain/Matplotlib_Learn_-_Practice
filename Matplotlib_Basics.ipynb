### Two options for subplots

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10,5))
# plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10,5))
# plot to each different index
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

### Which one should you use? (pyplot vs matplotlib vs OO method)
* When plotting something quickly, okay to use the pyplot method
* When plotting somthing more advanced, use the OO method

```python
hd.head()
```

```python
over_50 = hd[hd["age"]>50]
over_50.head()
```

```python
# Pyplot method
over_50.plot(kind='scatter',
           x='age',
           y='chol',
           c='target');
```

```python
over_50.target.values
```

```python
# OO method
fix, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
             x = 'age',
             y = 'chol',
             c= 'target',
             ax=ax);
#ax.set_xlim([45, 100])
```

## OO method from scartch

```python
fig, ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50['chol'],
                    c=over_50['target'])

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
      xlabel="Age",
      ylabel="Cholesterol");

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle='--');
```

```python
over_50.head()
```

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10),
                              sharex=True);

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"]);

# Customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
       # xlabel="Age", (for sharex, because xlabel is same for both plot)
       ylabel="Cholesterol");

# Add a legend ax0
ax0.legend(*scatter.legend_elements(),title="Target");

# Add a meanline
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--");

# ------------

# Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"]);

# Customize ax1
ax1.set(title="Heart Disease and Max Heart Rate",
       xlabel="Age",
       ylabel="Max Heart Rate");

# Add a legend ax0
ax1.legend(*scatter.legend_elements(),title="Target");

# Add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--");

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");

# Save the figure
fig.savefig("images/heart_disease_analysis.png");
```

```python
# Set the style
plt.style.use('seaborn-whitegrid')

fig, ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50['chol'],
                    c=over_50['target'],
                    cmap='winter') # This changes the color scheme

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
      xlabel="Age",
      ylabel="Cholesterol");

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle='--');
```

This plot shows some information about the heart disease dataset

```python
# Customizing the  x and y axis limitations

# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10),
                              sharex=True);

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap='winter');

# Customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
       # xlabel="Age", (for sharex, because xlabel is same for both plot)
       ylabel="Cholesterol");
# Changes x axis limts
ax0.set_xlim([50, 80])

# Add a legend ax0
ax0.legend(*scatter.legend_elements(),title="Target");

# Add a meanline
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--");

# ------------

# Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"],
                     cmap='winter');

# Customize ax1
ax1.set(title="Heart Disease and Max Heart Rate",
       xlabel="Age",
       ylabel="Max Heart Rate");
# Changes y axis limts
ax1.set_ylim([60, 200])

# Add a legend ax0
ax1.legend(*scatter.legend_elements(),title="Target");

# Add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--");

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");

# Save the figure
fig.savefig("images/heart_disease_analysis-plot.png");
```
